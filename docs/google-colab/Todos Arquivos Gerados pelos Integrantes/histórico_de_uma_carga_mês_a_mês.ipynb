```python
import pandas as pd

# Carregar os arquivos CSV
exportacoes2024 = 'https://balanca.economia.gov.br/balanca/bd/comexstat-bd/mun/EXP_2024_MUN.csv'
mun = 'https://balanca.economia.gov.br/balanca/bd/tabelas/UF_MUN.csv'
sh4 = 'https://balanca.economia.gov.br/balanca/bd/tabelas/NCM_SH.csv'


df = pd.read_csv(exportacoes2024, sep=";", encoding="latin1")
df_mun = pd.read_csv(mun, sep=";", encoding="latin1")
df_sh4 = pd.read_csv(sh4, sep=";", encoding="latin1")
```

```python

for mes in range(1,13):
  # Filtrar dados para o estado de São Paulo (SP)
  df_sp = df[df["SG_UF_MUN"] == "SP"]

  #filtrando para ser só do mês 1(janeiro)
  df_sp = df_sp[df_sp['CO_MES'] == mes]

  # Agrupar por município e SH4 e calcular a média do valor agregado
  municipios_agregado = df_sp.groupby(["CO_MUN", "SG_UF_MUN", "SH4","CO_MES"])['VL_FOB'].sum().reset_index()

  # Renomear coluna para compatibilidade com a base de municípios
  df_mun = df_mun.rename(columns={"CO_MUN_GEO": "CO_MUN"})

  # Mesclar com os nomes dos municípios
  municipios_agregado = municipios_agregado.merge(df_mun, on=["CO_MUN"], how="left")

  # Renomear coluna para compatibilidade com a base de produtos
  df_sh4 = df_sh4.rename(columns={"CO_SH4": "SH4", "NO_SH4_POR": "PRODUTO"})

  # Mesclar com os nomes dos produtos
  municipios_agregado = municipios_agregado.merge(df_sh4, on=["SH4"], how="left")

    #agrupando o nome do municipio e o mes e calculando a média do valor agregado
  municipios_agregado = municipios_agregado.groupby(['NO_MUN_MIN','CO_MES','PRODUTO'])['VL_FOB'].sum().reset_index()

  municipios_agregado = municipios_agregado[municipios_agregado['PRODUTO'] == 'Cavalos, asininos e muares, vivos']
  print(municipios_agregado[["NO_MUN_MIN", "PRODUTO", "VL_FOB"]])

  # Selecionar colunas relevantes e ordenar
  municipios_carga = municipios_agregado.sort_values(by="VL_FOB", ascending=False).head(10)

  # Exibir resultado (Se necessário Data Frame!) *** IMPORTANTE! ***
  print(municipios_carga[["NO_MUN_MIN", "PRODUTO", "VL_FOB"]])

  #gráfico
  ax = municipios_carga.plot(
      kind='bar',
      title=f'Carga(Cavalos, asininos e muares, vivos) no mês de {mes}',
      x='NO_MUN_MIN',
      y='VL_FOB',
      rot = 45,
      legend=False
  )

  ax.set_xlabel('Municípios')
  ax.set_ylabel('Valor FOB')

#salvando a imagem
  fig = plt.gcf()
  plt.savefig(f'/content/export_2024_mes{mes}.jpg', format='jpg', dpi=300, bbox_inches='tight')

  plt.show()


  # Salvar o DataFrame com as informações legíveis no CSV
  # municipios_top10.to_csv('/content/export_2013.csv', index=False)
```
